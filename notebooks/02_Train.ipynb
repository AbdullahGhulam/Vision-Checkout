```python
from ultralytics.data.converter import convert_coco
import os 

yolo_data_path = "../data/yolo"


if not os.path.exists(yolo_data_path):
    print(f" {yolo_data_path} not found. Starting conversion...")
    
    convert_coco(
        labels_dir='../data/', 
        save_dir=yolo_data_path,
        use_segments=False, 
        cls91to80=False,
    )
    
    print("✅ Conversion completed successfully!")
else:
    print(f"✅ {yolo_data_path} already exists. Skipping conversion.")
```

```python
from ultralytics import YOLO
model = YOLO("yolo11n.pt")
model.to('cuda')

model.train(
    data=r'C:\Users\taher\Vision-Checkout\data.yaml', 
    epochs=2, 
    imgsz=640, 
    batch=32, 
    device=0,
    workers=2
)
```
